# Checking for Null Values

```python
data.isnull().sum()
```

# a) drop all coulmns where there are more than 70% missing values 

```python

data.dropna(axis=1,thresh=0.70*len(data),inplace = True)
```

```python
data.head()
```

```python
data.shape
```

# b) lets define a subset of columns and drop records if all the values are NaN

```python
data.dropna(how='all', subset=['Email','CNIC','POBOX','Mobile'],inplace=True)
```

```python
data.shape
```

# c) lets define a subset of columns and drop records if any the values are NaN

```python
data.dropna(how='any', subset=['Email','CNIC','POBOX','Mobile'],inplace=True)
```

```python
# reset index
data.reset_index(drop=True,inplace=True)
```

# Changing to data types`

```python
# checking data types
data.dtypes
```

```python
# changing date of birth (DOB) to datetime
data['DOB']=pd.to_datetime(data['DOB'])
```

```python
# checking the change
data.dtypes
```

```python
# then we try to change pobox to int
data['POBOX']=pd.to_datetime(data['POBOX'])
```

```python
data
```

```python
# Pobox not converting in int so we do in another way
# validating Pobox (lenght of pobox should be 5 and it should not contan any alphabet)

cnt =0
for row  in data['POBOX']:
    if (len(str(row))!=5) or (any(c.isalpha() for c in row)):
        data.loc[cnt,'POBOX']=np.nan
    cnt+=1
```
